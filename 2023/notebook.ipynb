```python
def dists(text, rep = 3):
    btext = text == '.'
    h, w = btext.shape
    cols = []
    rows = []

    for i in range(h):
        if np.sum(btext[i, :]) == w:
            cols.append(i)

    for i in range(w):
        if np.sum(btext[:, i]) == h:
            rows.append(i)

    ntext = np.ones(text.shape)

    for i in range(h):
        if np.sum(btext[i, :]) == w:
            ntext[i,:] = rep

    for i in range(w):
        if np.sum(btext[:, i]) == h:
            ntext[:,i] = rep


    return ntext#.astype(int)
```

```python
def calc_dist(c1, c2, q):
    xi = [c1[0], c2[0]]
    yi = [c1[1], c2[1]]
    xi.sort()
    yi.sort()

    v = np.sum(q[xi[0]:xi[1], c1[1]])
    v += np.sum(q[c2[0], yi[0]:yi[1]])
    return v
```

```python
def hyperlink_markdown(path, done, begin = 1, end = 26):

    hyperlinks = ''

    for num in range(begin, end):
        if num in done:
            hyperlinks += f'[**]({path}{num:02d}.py) |'
        else:
            hyperlinks += f'[X]({path}{num:02d}.py) |'

    return hyperlinks

path = 'https://github.com/FrankNicoMol/AdventOfCode/blob/main/2023/day'
hyperlink_markdown(path, done = [1,2,3,4,6,7], begin = 1, end = 26)
```

```python
def table_formatter(begin = 1, end = 26):

    table_line = ''
    for  i in range(begin, end):
        table_line += f':-----:|'

    return table_line

table_formatter(1, 26)
```
