```python
dir_to_step = {'right': np.array([0, 1]),
               'left': np.array([0, -1]),
               'up': np.array([-1, 0]),
               'down': np.array([1, 0]),
               }

next_dir = {'right': {'right': 'down',
                      'straight': 'right',
                      'left': 'up',
                      },
            'left': {'right': 'up',
                     'straight': 'left',
                     'left': 'down',
                     },
            'down': {'right': 'left',
                     'straight': 'down',
                     'left': 'right',
                     },
            'up': {'right': 'right',
                   'straight': 'up',
                   'left': 'left',
                   },
            }
```

```python
def to_array(m):
    mlist = []
    for mi in m:
        mlist.append([*mi])

    return np.asarray(mlist)
```

```python
def calc_loss(block, pos, end_point, d, n_straight, visited):
    visited_from_here = visited.copy()
    h, w = block.shape

    visited_from_here[pos[0], pos[1]] = 1
    val = block[pos[0], pos[1]]

    if np.sum(pos != end_point) == 0:
        return val, visited_from_here
    else:
        v = []
        vfh = []
        dn = next_dir[d]['right']
        new_pos = pos + dir_to_step[dn]
        if 0 <= new_pos[0] < h and new_pos[1] >= 0 and new_pos[1] < w:

            if not visited[new_pos[0], new_pos[1]]:
                vt, vh = calc_loss(block, new_pos, end_point, dn, 0, visited_from_here)
                if vt != None:
                    v.append(vt)
                    vfh.append(vh)

        dn = next_dir[d]['left']
        new_pos = pos + dir_to_step[dn]
        if 0 <= new_pos[0] < h and new_pos[1] >= 0 and new_pos[1] < w:

            if not visited[new_pos[0], new_pos[1]]:
                vt, vh = calc_loss(block, new_pos, end_point, dn, 0, visited_from_here)
                if vt != None:
                    v.append(vt)
                    vfh.append(vh)

        if n_straight < 3:
            dn = next_dir[d]['straight']
            new_pos = pos + dir_to_step[dn]
            if 0 <= new_pos[0] < h and new_pos[1] >= 0 and new_pos[1] < w:

                if not visited[new_pos[0], new_pos[1]]:
                    vt, vh = calc_loss(block, new_pos, end_point, dn, n_straight + 1, visited_from_here)
                    if vt != None:
                        v.append(vt)
                        vfh.append(vh)

    if v == []:
        val = None
    else:

        val += np.min(v)
        visited_from_here = vfh[np.argmin(v)]

    return val, visited_from_here
```

```python
block = to_array(text).astype(int)
pos = np.array([0, 0])
loss = 0
end_point = np.array(block.shape) - np.array([1, 1])
n_straight = 0
dirs = ['right', 'down']
visited = np.zeros_like(block)
vs = []
for d in dirs:
    visited_from_here = visited.copy()
    val, visited_from_here = calc_loss(block, pos, end_point, d, n_straight, visited_from_here)
    if val:
        vs.append(val)
```

```python
def feas(p, h, w):
    feasible = True if (0 <= p[0] < h and p[1] >= 0 and p[1] < w) else False
    return feasible

```

```python
def get_graph(block):
    h, w = block.shape
    g = {}
    for i in range(h):
        for j in range(w):
            positions = []
            p1 = i - 1, j
            if feas(p1, h, w):
                positions.append([p1[0] * h + p1[1], block[p1[0], p1[1]], 'left'])
            p2 = i, j - 1
            if feas(p2, h, w):
                positions.append([p2[0] * h + p2[1], block[p2[0], p2[1]], 'up'])
            p3 = i + 1, j
            if feas(p3, h, w):
                positions.append([p3[0] * h + p3[1], block[p3[0], p3[1]], 'right'])
            p4 = i, j + 1
            if feas(p4, h, w):
                positions.append([p4[0] * h + p4[1], block[p4[0], p4[1]], 'down'])
            g[i*h+j] = positions
    return g
```

```python
# takes the graph and the starting node
# returns a list of distances from the starting node to every other node
def naive_dijkstras(graph, root):
    n = len(graph)
    # initialize distance list as all infinities
    dist = [Inf for _ in range(n)]
    # set the distance for the root to be 0
    dist[root] = 0
    # initialize list of visited nodes
    visited = [False for _ in range(n)]
    # loop through all the nodes
    for _ in range(n):
        # "start" our node as -1 (so we don't have a start node yet)
        u = -1
        # loop through all the nodes to check for visitation status
        for i in range(n):
            # if the node 'i' hasn't been visited and
            # we haven't processed it or the distance
            #
            # dist[u]):
                u = i
        # all the nodes have been visited or we can't reach this node
        if dist[u] == Inf:
            break
        # set the node as visited
        visited[u] = True
        # compare the distance to each node from the "start" node
        # to the distance we currently have on file for it
        for v, l in graph[u]:
            if dist[u] + l < dist[v]:
                dist[v] = dist[u] + l
    return dist
```

```python
def get_sets(p):
    string = ''.join(p)
    q = str(string).split('.')
    lens = [len(qi) for qi in q if len(qi) != 0]
    return lens


def rec(p, n, qms, c):
    global tdict

    if c != len(qms):
        gsets = get_sets(p[:qms[c]])
        #print(gsets, p[:qms[c]])
        if str(gsets) + ',' + str(c) in tdict.keys():
            if c != 0:
                return tdict[str(gsets) + ',' + p[qms[c - 1]] + ',' + str(c)]
            else:
                return tdict[str(gsets) + ',' + str(c)]
        else:
            t1 = p.copy()
            t1[qms[c]] = '#'
            t2 = p.copy()
            t2[qms[c]] = '.'

            a1 = rec(t1, n, qms, c + 1)
            a2 = rec(t2, n, qms, c + 1)
            val = a1 + a2
            if c != 0:
                key = str(gsets) + ',' + p[qms[c - 1]] + ',' + str(c)
            else:
                key = str(gsets) + ',' + str(c)
            tdict[key] = val
            return val
    else:
        if np.array_equal(np.array(get_sets(p)), n):
            key = str(get_sets(p)) + ',' + p[qms[c - 1]] + ',' + str(c)
            tdict[key] = 1
            return 1
        else:
            key = str(get_sets(p)) + ',' + p[qms[c - 1]] + ',' + str(c)
            tdict[key] = 0
            return 0


def get_line_val(line, second):
    global tdict
    tdict = {}

    v, n = line.split(' ')
    n = np.asarray(n.split(',')).astype(int)
    if second:
        v = '?'.join([v] * 5)

    n = list(n) * 5 if second else list(n)
    qms = np.where(np.array([*v]) == '?')

    total = rec(np.array([*v]), n, qms[0], 0)

    return total


def get_value(text, second):
    val = np.sum([get_line_val(line, second) for line in text])

    return val
```

```python
def hyperlink_markdown(path, done, begin=1, end=26):
    hyperlinks = ''

    for num in range(begin, end):
        if num in done:
            hyperlinks += f'[**]({path}{num:02d}.py) |'
        else:
            hyperlinks += f'[X]({path}{num:02d}.py) |'

    return hyperlinks


path = 'https://github.com/FrankNicoMol/AdventOfCode/blob/main/2023/day'
hyperlink_markdown(path, done=[1, 2, 3, 4, 6, 7], begin=1, end=26)
```
