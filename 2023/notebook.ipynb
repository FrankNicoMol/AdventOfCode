```python
def get_sets(p):
    string = ''.join(p)
    q = str(string).split('.')
    lens = [len(qi) for qi in q if len(qi) != 0]
    return lens


def rec(p, n,qms,  c):
    global tdict

    if c != len(qms):
        gsets =get_sets(p[:qms[c]])
        #print(gsets, p[:qms[c]])
        if str(gsets) + ',' + str(c)in tdict.keys():
            if c != 0:
                return tdict[str(gsets) + ',' + p[qms[c-1]] + ',' + str(c)]
            else:
                return tdict[str(gsets) + ',' + str(c)]
        else:
            t1 = p.copy()
            t1[qms[c]] = '#'
            t2 = p.copy()
            t2[qms[c]] = '.'

            a1 = rec(t1, n,qms, c+1)
            a2 = rec(t2, n,qms, c+1)
            val = a1 + a2
            if c!= 0:
                key = str(gsets) + ',' + p[qms[c-1]] + ',' + str(c)
            else:
                key = str(gsets) + ',' + str(c)
            tdict[key] = val
            return val
    else:
        if np.array_equal(np.array(get_sets(p)), n):
            key = str(get_sets(p)) + ',' + p[qms[c-1]] + ',' + str(c)
            tdict[key] = 1
            return 1
        else:
            key = str(get_sets(p)) + ',' + p[qms[c-1]] + ',' + str(c)
            tdict[key] = 0
            return 0

def get_line_val(line, second):
    global tdict
    tdict = {}

    v, n = line.split(' ')
    n = np.asarray(n.split(',')).astype(int)
    if second:
        v = '?'.join([v] * 5)

    n = list(n) * 5 if second else list(n)
    qms = np.where(np.array([*v]) == '?')


    total = rec(np.array([*v]), n, qms[0], 0)


    return total


def get_value(text, second):
    val = np.sum([get_line_val(line, second) for line in text])

    return val
```

```python
def hyperlink_markdown(path, done, begin = 1, end = 26):

    hyperlinks = ''

    for num in range(begin, end):
        if num in done:
            hyperlinks += f'[**]({path}{num:02d}.py) |'
        else:
            hyperlinks += f'[X]({path}{num:02d}.py) |'

    return hyperlinks

path = 'https://github.com/FrankNicoMol/AdventOfCode/blob/main/2023/day'
hyperlink_markdown(path, done = [1,2,3,4,6,7], begin = 1, end = 26)
```

```python
def table_formatter(begin = 1, end = 26):

    table_line = ''
    for  i in range(begin, end):
        table_line += f':-----:|'

    return table_line

table_formatter(1, 26)
```
