```python
def binary_to_string(bin):
    key = ['.', '#']
    new = [key[int(i)] for i in bin]
    return ''.join(new)
```

```python
def get_sets(string):
    q = str(string).split('.')
    lens = [len(qi) for qi in q if len(qi) != 0]
    return lens
```

```python
import itertools
```

```python
def get_line_val(line):
    v, n = line.split(' ')
    n = np.asarray(n.split(',')).astype(int)
    n = list(n)
    qms = np.where(np.array([*v]) == '?')
    reps = len(qms[0])
    lst = list(itertools.product([0, 1], repeat=reps))
    array =  np.zeros(np.array([*v]).shape) + (np.array([*v]) == '#')
    total = 0
    for li in lst:
        array[qms] = np.array(li)
        set_c = get_sets(binary_to_string(array))
        if set_c == n:
            total += 1
    return total
```

```python
def hyperlink_markdown(path, done, begin = 1, end = 26):

    hyperlinks = ''

    for num in range(begin, end):
        if num in done:
            hyperlinks += f'[**]({path}{num:02d}.py) |'
        else:
            hyperlinks += f'[X]({path}{num:02d}.py) |'

    return hyperlinks

path = 'https://github.com/FrankNicoMol/AdventOfCode/blob/main/2023/day'
hyperlink_markdown(path, done = [1,2,3,4,6,7], begin = 1, end = 26)
```

```python
def table_formatter(begin = 1, end = 26):

    table_line = ''
    for  i in range(begin, end):
        table_line += f':-----:|'

    return table_line

table_formatter(1, 26)
```
