```python
def read_map(file):
    text = read_txt(file)
    dirs = text[0]
    maps = text[2:]

    maps_list = {}
    for m in maps:
        map_split = m.split('=')
        maps_list[map_split[0].strip()] = [map_split[1].split(',')[0].strip()[1:],
                                           map_split[1].split(',')[1].strip()[:-1]]

    return dirs, maps_list


def calculate_steps(file, second):
    dirs, maps = read_map(file)
    dest_map = next(iter(maps))
    steps = 0
    while dest_map != 'ZZZ':
        print(dest_map, maps[dest_map])
        dest_map = maps[dest_map][(dirs[steps % len(dirs)] == 'R')]
        steps += 1
    return steps
```

```python
def hyperlink_markdown(path, done, begin = 1, end = 26):

    hyperlinks = ''

    for num in range(begin, end):
        if num in done:
            hyperlinks += f'[**]({path}{num:02d}.py) |'
        else:
            hyperlinks += f'[X]({path}{num:02d}.py) |'

    return hyperlinks

path = 'https://github.com/FrankNicoMol/AdventOfCode/blob/main/2023/day'
hyperlink_markdown(path, done = [1,2,3,4,6,7], begin = 1, end = 26)
```

```python
def table_formatter(begin = 1, end = 26):

    table_line = ''
    for  i in range(begin, end):
        table_line += f':-----:|'

    return table_line

table_formatter(1, 26)
```
