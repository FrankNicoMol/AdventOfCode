```python
from tkinter import Tk
import time

def clipboard(text):
    tk_object = Tk()
    tk_object.withdraw()
    tk_object.clipboard_clear()
    tk_object.clipboard_append(text)
    time.sleep(.3)
    tk_object.update()
    tk_object.destroy()


def get_md_string(years, n_days = 25):
    md_string = ''
    for i in range(1, n_days+1):
        md_string += '| '
        for year in years:
            md_string += f'[Day {i:02d}](https://adventofcode.com/{year}/day/{i}) - [-](https://github.com/FrankNicoMol/AdventOfCode/blob/main/{year}/day{i:02d}.py) |'
        md_string += '\n'
    clipboard(md_string)
```

```python
get_md_string(years = [2024, 2023])
```
