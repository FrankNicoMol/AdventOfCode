```python
from tkinter import Tk
import time


def clipboard(text):
    tk_object = Tk()
    tk_object.withdraw()
    tk_object.clipboard_clear()
    tk_object.clipboard_append(text)
    time.sleep(.3)
    tk_object.update()
    tk_object.destroy()


def get_md_string(years, stars_years, n_days=25):
    md_string = ''
    for i in range(1, n_days + 1):
        md_string += '| '
        for year, stars_year in zip(years, stars_years):
            stars = ' :x: '
            stars_day = stars_year[i-1]
            if stars_day > 0:
                stars = f' {":star:"*stars_day} '
            md_string += f'[Day {i:02d}](https://adventofcode.com/{year}/day/{i}):  [{stars}](https://github.com/FrankNicoMol/AdventOfCode/blob/main/{year}/day{i:02d}.py) |'
        md_string += '\n'
    clipboard(md_string)
```

```python
def stars_year(year, day01=0, day02=0, day03=0, day04=0, day05=0, day06=0, day07=0, day08=0, day09=0, day10=0, day11=0,
               day12=0, day13=0, day14=0, day15=0, day16=0, day17=0, day18=0, day19=0, day20=0, day21=0, day22=0,
               day23=0, day24=0, day25=0):
    return [day01, day02, day03, day04, day05, day06, day07, day08, day09, day10, day11,
            day12, day13, day14, day15, day16, day17, day18, day19, day20, day21, day22,
            day23, day24, day25]
```

```python
stars_2022 = stars_year(2022, day01=2, day02=2)
stars_2023 = stars_year(2023, day01=2, day02=2, day03=2, day04=2, day05=0, day06=2, day07=2, day08=2, day09=2, day10=2, day11=2,
               day12=2, day13=2, day14=2, day15=2, day16=2)
stars_2024 = stars_year(2024)

stars_years = [stars_2024, stars_2023, stars_2022]
```

```python
get_md_string(years=[2024, 2023, 2022], stars_years = stars_years )
```
